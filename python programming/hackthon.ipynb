```python

import random
def generatemarksdata(n, filename):
    with open(filename,'w') as marksfile:
        for i in range(0,n):
            marks=random.randint(0,101)
            marksfile.write(str(marks)+'\n')
    return
generatemarksdata(2000,'marks1data.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def readfiledata(filename):
    with open (filename,'r') as marksfile:
        sum=0
        c=d=e=f=g=0
        a=[]
        for i in range(0,2000):
            a.append(int(marksfile.readline()))
        for i in range (0,len(a)):
            if a[i]>=80:
                c=c+1
            if 60<=a[i]<80:
                d=d+1
            if 50<=a[i]<60:
                e=e+1
            if 40<=a[i]<50:
                f=f+1
            if a[i]<40:
                g=g+1

    labels = 'Distinction', 'First Class', 'Second Class', 'Third Class','Fail'
    sizes = [c,d,e,f,g]
    colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','gray']
    explode = (0.1, 0, 0, 0, 0)  
    plt.pie(sizes, explode=explode, labels=labels, colors=colors,
    autopct='%1.1f%%', shadow=True, startangle=140)

    plt.axis('equal')
    plt.show()
readfiledata('marks1data.csv')
```
