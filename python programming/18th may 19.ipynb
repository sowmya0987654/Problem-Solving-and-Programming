# problem 2 :
## problem statement
drawing time-voltasge graph using matplotlib

## constraint


## test cases
* test case 1
* test case 2
* test case 3

```python
import numpy as np
t=np.arange(0.0,2.0,0.1)
v=1+np.sin(2 * np.pi * t)
plt.xlabel('time')
plt.ylabel('voltage')
plt.grid(True)
plt.plot(t,v)
plt.savefig('Figures/time-voltage.png')
plt.show()
```

# problem 3:
## problem statement
plot a line graph on **income vs **year for the state of **california

### constraint


### test cases
* test case 1
* test case 2
* test case 3

```python
import pandas as pd
df=pd.read_csv('DataFiles/income.csv')
y=df.values[4]
x=df.columns
```

```python
years=[]
for i in range(2,len(x)):
    years.append(x[i])
years
income=[]
for i in range(2,len(y)):
    income.append(y[i])
income


plt.xlabel('years')
plt.ylabel('income')
plt.title('income of california state')
plt.plot(years,income)
plt.grid(True)

plt.show()
```

```python
width=1/2
plt.title('income of california')
plt.bar(years,income,width,color='gray')
plt.savefig('Figures/line-graph-california')
```

```python
n=50
x=np.random.rand(n) #generate 50 random numbers
y=np.random.rand(n) #generate 50 random numbers
colorlist=np.random.rand(n) #generate 50 random color values
arealist=np.pi * (15 * np.random.rand(50))**2 #generate 50 rtandom area of circle
plt.scatter(x,y,s=arealist,c=colorlist,alpha=0.5)
plt.savefig('Figures/scatter-picture')
```
